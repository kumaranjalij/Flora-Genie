## Combining

```python
# def combine_columns(new_data, column1, column2, new_column):

#     # Set the values of the new column based on the condition
#     new_data[new_column] = (new_data[column1] | new_data[column2]).astype(int)
    
#     # Drop the original columns
#     new_data.drop(columns=[column1, column2], inplace=True)
    
#     return new_data

def combine_columns(new_data, column1, column2, new_column):

    # Calculate the OR value of the two columns
    or_value = new_data[column1] | new_data[column2]
    
    # Update the new column only if the OR value is 1
    new_data[new_column] = or_value.mask(or_value != 1, new_data[new_column])
    
    # Drop the original columns
    new_data.drop(columns=[column1, column2], inplace=True)
    
    return new_data
```

```python
# The function will return the modified DataFrame
new_data = combine_columns(new_data, 'Water_Can tolerate less', 'Water_Requires less', 'Water_Requires_less')

# The function will return the modified DataFrame
new_data = combine_columns(new_data, 'Water_Can tolerate more', 'Water_Requires more', 'Water_Requires_more')

# Display the modified DataFrame
new_data
```
